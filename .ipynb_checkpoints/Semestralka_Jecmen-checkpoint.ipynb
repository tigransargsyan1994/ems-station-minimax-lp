```python
from pulp import *
import pandas as pd
import numpy as np
```

```python
# Import vstupnich dat
data = pd.read_excel('Data.xlsx', index_col = 0, sheet_name = 0)
naklady = pd.read_excel('Data.xlsx', index_col = 0, sheet_name = 1)
```

```python
data
```

```python
vrcholy = data.index.to_list()
print('vrcholy =', data.index.to_list())
```

```python
f = naklady['Náklady'].to_dict()
print('f = ', naklady['Náklady'].to_dict())
```

```python
a = data.to_dict()
print('a = ', data.to_dict())
```

```python
# Model
model = LpProblem("Pokryvaci_uloha", LpMinimize)

kombinace = [(i,j) for i in vrcholy for j in vrcholy]

# Deklarace proměnných
x = LpVariable.dicts("x",(vrcholy, vrcholy),0 , cat = 'Binary')
y = LpVariable.dicts("y",(vrcholy),0 , cat = 'Binary')
z = LpVariable("z",0 , cat = 'Binary')
q = LpVariable("q",0 , cat = 'Integer')

# Ucelova funkce
model += q

# Omezující podmínky
for j in vrcholy:
    model += lpSum(x[i][j] * a[i][j] for i in vrcholy) <= q
    model += lpSum(x[i][j] for i in vrcholy) >= 1
    
for i in vrcholy:
    model += lpSum(x[i][j] for j in vrcholy) <= 1000 * y[i]

model += lpSum(y[i] * f[i] for i in vrcholy) <= 50

model += y['v2'] <= z
model += y['v6'] <= (1 - z)

# Výpis výsledků
model.solve()
print("Status:", LpStatus[model.status])

for v in model.variables():
    if v.varValue > 0:
        print(v.name, "=", v.varValue)
        
print("Maximální dojezdová doba je: ", value(model.objective)) 
```
